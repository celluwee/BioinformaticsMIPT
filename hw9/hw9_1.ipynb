Для следующих данных:

79  80  80  80  74  80  80  79  64  78  73  78  74  45  81  48  80  82  82  70

```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as sps
import matplotlib.pyplot as plt
```

```python
size = 10000
sample = sps.uniform.rvs(size=size)
```

```python
def draw_quantiles(data, alphas, ymax=0.4, title=None, bins=20):

    f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)},figsize=(14, 8))
    f.suptitle(f"Распределение {title}")

    sns.boxplot(x=data, ax=ax_box, color="steelblue")

    for alpha in alphas:
        nearest = np.nanquantile (data, alpha)
        ax_hist.vlines(nearest, 0, ymax, color="red", label=f'{alpha} квантиль')
    sns.histplot(data=data, kde=True, ax=ax_hist, stat='density', bins=bins)

    plt.legend()
    plt.show()
```

```python
alphas = np.array([0.25, 0.5, 0.75])
```

```python
draw_quantiles(sample, alphas, ymax=1.4, title='равномерное')
```

```python
arr = np.array([79, 80, 80, 80, 74, 80, 80, 79, 64, 78, 73, 78, 74, 45, 81, 48, 80, 82, 82, 70])
```

```python
Q1 = np.quantile(arr, 0.25)
Q2 = np.quantile(arr, 0.5)
Q3 = np.quantile(arr, 0.75)
IQR = Q3 - Q1
outliers = arr[~((Q1 - 1.5*IQR < arr) & (arr < Q3 + 1.5*IQR))]
```

```python
d = {'min': min(arr), 'Q25':Q1, 'медиана Q5':Q2,'Q75':Q3,'межквартильный размах':IQR,'max':max(arr)}
```

```python
pd.DataFrame(data=d, index=['arr'])
```

```python
print('Потенциальные выбросы:', list(outliers))
```

```python
draw_quantiles(arr, alphas, ymax=0.2, title='manual')
```
